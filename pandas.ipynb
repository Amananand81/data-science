pandas- pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. it provides two primary data structure.


# pandas 1-d

```python
import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print(series)

data2=pd.array([10,20,30,40,50])
print(data2)

data3=pd.array([10,20,30,40,50])
print(data3)
```

pandas 1-d dictinory...

```python

data={'a':1,'b':2,'c':3}
series=pd.Series(data)
print(series)
```

# manipulating data with dataframe

```python
# add a new column to the dataframe
df['salary']=[50000,60000,70000]
df
```
